```python
def get_data_local_csv():

    """
    Chose, import, clean/transform data from a local CSV file.

    This function gives the user options for input, takes that input and reads data from the corrosponding local .csv file. It then calls the clean_local_csv() function.
    The user only gets three oprtunities to enter a valid stock symbol from the list provided and then terminates. Valid symbols can be entered in lower, upper or mixed case.

    Parameters
    ----------
    Valid stock symbol.

    Returns
    -------
    VAR symbol

    Examples
    --------
    MSFT, AAPL
    """ 


    for _ in range(3):
        symbols = ['AAPL', 'AMD', 'AMZN', 'CSCO', 'META', 'MSFT', 'NFLX', 'QCOM', 'SBUX', 'TSLA']
        symbol = str(input("Please enter one of the stock symbols listed here!" + str(symbols)).upper())

        if symbol in symbols:
            clean_local_csv(symbol= symbol)
               
            
        else:
            print('You have entered an invalid symboll! Please try again!')
            pass

    print("Function terminates after 3 invalid entries.")




#def get_csv_api():

    #import csv
    #import requests

    #CSV_URL = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol=IBM&interval=60min&slice=year1month1&apikey=E5CX4VOWGQH169RE'

    #with requests.Session() as s:
        #download = s.get(CSV_URL)
        #decoded_content = download.content.decode('utf-8')
        #cr = csv.reader(decoded_content.splitlines(), delimiter=',')
        #my_list = list(cr)
        #for row in my_list:
            #print(row)

        
        
def clean_local_csv(symbol):
    """
    Clean/transform data from a local CSV file.

    This function is specifically written to read data from a local .csv file which has been manually downloaded from which corrosponds to the user input. It subsequently cleans/transforms the data into a pandas dataframe. 
    The user only gets three oprtunities to enter a valid stock symbol from the list provided and then terminates. Valid symbols can be entered in lower, upper or mixed case.

    Parameters
    ----------
    Valid stock symbol.

    Returns
    -------
    pandas dataframe.

    Examples
    --------
    MSFT, AAPL
    """ 

    import pandas as pd

    df = pd.read_csv('/Users/miked/Downloads/HistoricalData_' + symbol + '.csv')
    df['Symbol'] = symbol
    df['Open'] = df['Open'].str.replace("$", '', regex=True).astype(float)
    df['High'] = df['High'].str.replace("$", '', regex=True).astype(float)
    df['Low'] = df['Low'].str.replace("$", '', regex=True).astype(float)
    df['Close/Last'] = df['Close/Last'].str.replace("$", '', regex=True).astype(float)
    df.rename(columns={'Close/Last': 'Close'}, inplace=True)
    df['Date']= pd.to_datetime(df['Date'])
        
    return df
    


```
